```python
import chess
import chess.pgn
```

```python
with open('2008_SCT_LadiesOpen.pgn') as pgn:
# with open('../train/lichess_db_standard_rated_2017-02.pgn') as pgn:
    pgn.seek(0, 2)
    file_size = pgn.tell()
    pgn.seek(6547)
    offsets = []
    count = 0
    print(chess.pgn.read_headers(pgn))
    while chess.pgn.read_headers(pgn):
        count = count + 1
        offset = pgn.tell()
        offsets.append(offset)
        print(f'{count}: {offset} ({offset / file_size * 100.0}%)')

    for index, offset in enumerate(offsets):
        pgn.seek(offset)
        print(f'{index}: {offset}')
```
